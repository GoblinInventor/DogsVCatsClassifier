```python
import numpy as np
def randomInBounds(x1,x2,z):
    Y = np.random.random(z)
    Yp = np.ones((z))
    Yp = Yp*x1
    scale = abs(x2-x1)
    Y = Y*scale
    Y = Y + Yp
    return Y.astype('int')

def drawSquare(s,x,y,a):
    sq = np.zeros((s,s))
    for i in range(a):
        for j in range(a):
            sq[x+i][y+j] = 1
    return sq

def drawTriangle(s,x,y,a):
    sq = np.zeros((s,s))
    for i in range (a):
        for j in range(a):
            if j < i:
                sq[x+i][y+j] = 1
    return sq

def drawTrainingExamples(IMAGE_SIZE,SPECIMEN_COUNT):
    MAXLEN = IMAGE_SIZE/4
    MINLEN = IMAGE_SIZE/10
    x_centers = randomInBounds(IMAGE_SIZE/4,3*IMAGE_SIZE/4,SPECIMEN_COUNT)
    y_centers = randomInBounds(IMAGE_SIZE/4,3*IMAGE_SIZE/4,SPECIMEN_COUNT)
    side_len = randomInBounds(MINLEN,MAXLEN,SPECIMEN_COUNT)
    myshapes = []
    for i in range(SPECIMEN_COUNT/2):
        #def drawSquare(s,x,y,a):
        Zs = drawSquare(IMAGE_SIZE,x_centers[i],y_centers[i],side_len[i])
        Zt = drawTriangle(IMAGE_SIZE,x_centers[i+1],y_centers[i+1],side_len[i+1])
        s = Zs,1
        t = Zt,0
        myshapes.append(s)
        myshapes.append(t)
    return myshapes
trainingset = drawTrainingExamples(40,300)
testingset = drawTrainingExamples(40,100)

print len(trainingset)
print len(trainingset[0])
print trainingset[0][1]
print trainingset[0][0]
def nonlin(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))


X = []
y = []
for i in range(len(trainingset)):
    entry = []
    for j in range(40):
        for k in range(40):
            entry.append(trainingset[i][0][j][k])
    X.append(entry)
    y.append(trainingset[i][1])

#print "y.shape"
#print y.shape
#print len(X[0])
print len(X)
print len(X[0])
Xt = np.asarray(X)
yt = np.asarray([y])
yt = yt.T
print Xt.shape
print yt.shape
syn0 = np.random.random((40*40,40*80))-1
syn1 = np.random.random((40*80,1))-1
print "Beginning Training"
for iter in range(10000):
    l0 = Xt
    l1 = nonlin(np.dot(l0,syn0))
    l2 = nonlin(np.dot(l1,syn1))
    l2_error = yt-l2
    if (iter%1000)==0:
        print "Error: " + str(np.mean(np.abs(l2_error)))
    l2_delta = l2_error*nonlin(l2,deriv=True)
    l1_error = l2_delta.dot(syn1.T)
    l1_delta = l1_error * nonlin(l1,deriv=True)
    syn1 += l1.T.dot(l2_delta)
    syn0 += l0.T.dot(l1_delta)
print "Output After Training: "
#getcontext().prec = 3
for i in range(len(l2)):
    method = "%.6f" % l2[i]
    method2 = "%.6f" % y[i][0]
    print method + " | " + method2
```

```python
Xn = []
yn = []

for i in range(len(testingset)):
    entry = []
    for j in range(40):
        for k in range(40):
            entry.append(testingset[i][0][j][k])
    Xn.append(entry)
    yn.append(testingset[i][1])

l0 = Xn
l1 = nonlin(np.dot(l0,syn0))
l2 = nonlin(np.dot(l1,syn1))

diff = l2-yn
print "Error: " + str(np.mean(np.abs(diff)))
print "Standard Deviation: " + str(np.std(np.abs(diff)))
diffsclose = 0
misclass = 0
for i in range(len(l2)):
    method = "%.6f" % l2[i]
    method2 = "%.6f" % yn[i]
    diffval = "%.6f" % (l2[i]-yn[i])
    print method + " | " + method2 + " | " + diffval  
    if (abs(l2[i]-yn[i]) < 0.05):
        diffsclose = diffsclose+1
    if (abs(l2[i]-yn[i]) > 0.5):
        misclass = misclass + 1
diffsin = "%i" % diffsclose
miscall = "%i" % misclass
print "close vals: " + diffsin
print "misclassified: " + miscall
```
